```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/students.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['result'].value_counts()
```

```python
sns.countplot(x='result', data=df)
plt.title("Pass vs Fail Distribution")
plt.show()
```

```python
sns.boxplot(x='result', y='study_hours', data=df)
plt.title("Study Hours vs Result")
plt.show()
```

```python
sns.boxplot(x='result', y='attendance', data=df)
plt.title("Attendance vs Result")
plt.show()
```

```python
sns.boxplot(x='result', y='previous_marks', data=df)
plt.title("Previous Marks vs Result")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```
